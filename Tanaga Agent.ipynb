## Handle Interactions

This function handles user interactions with the LLM using ChatML, with each messaging expected to have a system, user, or assistant.

```python
def generate_response(messages: List[Dict]) -> str:
   """Call LLM to get response using Deepseek."""
   # Retrieve Deepseek API key from Colab secrets and set environment variable
   deepseek_api_key = userdata.get('DEEPSEEK_API_KEY')
   client = OpenAI(api_key=deepseek_api_key, base_url="https://api.deepseek.com")

   response = client.chat.completions.create(
      model="deepseek-reasoner", # Specify the Deepseek model
      messages=messages,
      max_tokens=1024,
      stream=False
   )
   return response.choices[0].message.content

def list_files() -> List[str]:
    """List all files in the current directory."""
    return [f for f in os.listdir('.') if os.path.isfile(f)]

def read_file(file_name: str) -> str:
    """Read the content of a file."""
    with open(file_name, 'r') as f:
        return f.read()
```

```python
agent_rules = [{
    "role": "system",
    "content": """
Write a Tanaga, a Filipino style of poetry, written in the form of four lines written in an A-A-B-B format, where each letter represents a line, and seven total syllables in each line.

With this context, use the following guidelines:

- **Structure**: The poem must have exactly four lines.
- **Syllables**: Each line should consist of exactly seven syllables in both English and Tagalog.
- **Language**: Respond in either English or Tagalog, as specified by the user's request.
  - If the language of the prompt differs from the target language, first translate the prompt before creating the poem.

# Steps

1. **Translation**: If necessary, translate the prompt into the target language.
2. **Analysis**: Understand the central theme or concept from the prompt.
3. **Composition**: Write a meaningful and creative poem adhering strictly to the seven-syllables-per-line rule.

# Output Format

Provide the poem in plain text, adhering to the above rules. Ensure each line has exactly seven syllables, and the overall tone matches the theme of the prompt.

# Examples

- **Input**: 'Write a poem about a sleepless night in Tagalog.'
  - **Output**:
    - 'Ma ingay dito ngayon,
    - Nag-uusap pa sila,
    - Pero gabi na ngayon,
    - Di ako makatulog.'

- **Input**: 'Write a poem about a sleepless night in English.'
  - **Output**:
    - 'It's so loud this evening.
    - All of their conversations,
    - Last for hours at nighttime.
    - I can never sleep again.'

# Notes

- Comprehend if the request is for a tanaga in English or Tagalog.
- Accuracy in syllable count is crucial; ensure each line is checked carefully to ensure that only seven syllables are written per line.
- Simplify complex themes before beginning the poem.
- Aim for simplicity and clarity in the poem's structure and language.
"""}]
```

```python
memory = [
    {"role": "system", "content": "Keep responses within 300 words total."},
]
```

## Implementation

Similarly to ChatGPT, I prompted DeepSeek for two tanagas that integrate the winter season and my upcoming graduation, one in English and the other in Tagalog. The related code and results are the following:

```python
print("Hello, I am a tanaga agent who can help you write a tanaga, which is a traditional Tagalog poem.")

user_prompt = input("What would you like to write a tanaga about? ")

# Add user's new query to memory
memory.append({"role": "user", "content": user_prompt})

# 1. Construct prompt: Combine agent rules with current memory.
prompt = agent_rules + memory

# 2. Generate response from LLM
print("Agent thinking...")
response = generate_response(prompt)
print(f"Agent response: {response}")

# 3. The LLM's response directly adheres to the agent_rules format.

# 4. Update memory with the LLM's response
memory.append(
    {"role": "assistant", "content": response}
)
```

```python
print("Hello, I am a tanaga agent who can help you write a tanaga, which is a traditional Tagalog poem.")

user_prompt = input("What would you like to write a tanaga about? ")

# Add user's new query to memory
memory.append({"role": "user", "content": user_prompt})

# 1. Construct prompt: Combine agent rules with current memory.
prompt = agent_rules + memory

# 2. Generate response from LLM
print("Agent thinking...")
response = generate_response(prompt)
print(f"Agent response: {response}")

# 3. The LLM's response directly adheres to the agent_rules format.

# 4. Update memory with the LLM's response
memory.append(
    {"role": "assistant", "content": response}
)
```

## ChatGPT vs. DeepSeek Comparison

Though I used a single prompt and a sample size of 1, similarities and differences are present in the outputs by ChatGPT and DeepSeek to these prompts.

ChatGPT's English response matched the A-A-B-B topical style of the tanaga and kept each line to around seven syllables (7-7-7-7-7-7-7-7), but it wrote too much content. Meanwhile, the Tagalog version looks more like a tanaga (in terms of structure), with both the A-A-B-B style and kept each line to around seven syllables (7-7-7-8).

DeepSeek's English response was also around seven syllables for each line (7-8-7-8) and kept the A-A-B-B style. The Tagalog response also kept to around seven syllables per line (7-7-8-7) and kept the A-A-B-B.

For this  attempt to prompt each model in different languages, it appears that ChatGPT's GPT-4.1 model and DeepSeek's reasoner model performed similarly for the Tagalog response in terms of structure and given the same system constraints.

However, deepseek-reasoner's output was far more interesting as the tanaga seemed to flow consistently from the imagery of winter and made an implicit connection to my "new life" as a result of earning my master's degree. Meanwhile, GPT-4.1 was descriptive without inference, though it also flowed well. But I acknowledge that I may just have more animated expectations and that others may find GPT-4.1's subtle output to be preferable.

Similarly, deepseek-reasoner's output for the English tanaga was structurally more preferable (with a 7-8-7-8 pattern) and, in far fewer words than GPT-4.1's output, used nature imagery effectively to encapsulate this new period in my life. Meanwhile, GPT-4.1 was more simple and direct, but added value with its extra lines, even if it isn't a "proper" tanaga.

Compare this to GPT 4.1's uncached input cost of \$2.00 per million tokens and cached input cost of \$0.50 per million tokens. Meanwhile, output costs are \$8.00 and \$0.42 per million tokens for GPT 4.1 and deepseek-reasoner, respectively. This leads to an overall cost of $10.50 per million tokens for GPT 4.1 and an overall cost of \$2.88 per million tokens for deepseek-reasoner. Finally, GPT 4.1 was more time efficient and spent less time to yield responses compared to deepseek-reasoner.

## Recommendations

For casual users who just want inspiration, DeepSeek is, by far, the more cost-effective option, even if time may not be on their side. However, time-sensitive applications (unlike generating tanagas) may require GPT-4.1 (and better models') speed. For a cheaper reasoning model for creative applications, like writing a tanaga or getting advice, it may be also worth it to wait a little bit to yield a response. Finally, it should also be considered that longer response times also require more energy on the server-side.
